```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
#Reading the data
train=pd.read_csv("AirplaneAccident.csv")
```

```python
print(train.info())
```

```python
train = train.sample(frac=1)
```

```python
train['Severity'].value_counts().plot.bar(figsize=(8,8),title='Severity')
plt.xlabel('Severity')
plt.ylabel('Count')
```

```python
#Visualization of train set 
#Feature Variables
#Histogram
f, ax = plt.subplots(2,5, figsize=(17, 8))
sns.distplot(train['Safety_Score'],ax = ax[0,0], kde = False)
sns.distplot(train['Days_Since_Inspection'],ax = ax[0,1], kde = False)
sns.distplot(train['Total_Safety_Complaints'],ax = ax[0,2], kde = False)
sns.distplot(train['Control_Metric'],ax = ax[0,3], kde = False)
sns.distplot(train['Turbulence_In_gforces'],ax = ax[0,4], kde = False)
sns.distplot(train['Violations'],ax = ax[1,0], kde = False)
sns.distplot(train['Cabin_Temperature'],ax = ax[1,1], kde = False)
sns.distplot(train['Accident_Type_Code'],ax = ax[1,2], kde = False)
sns.distplot(train['Max_Elevation'],ax = ax[1,3], kde = False)
sns.distplot(train['Adverse_Weather_Metric'],ax = ax[1,4], kde = False)
```

```python
target = "Severity"
X = train.drop(columns=[target])
Y = train[target]
#determining the importance of the features
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(X, Y)
model.feature_importances_
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.sort_values(ascending=True).plot(kind='barh',cmap='viridis')
plt.show()
```

```python
#Selecting relevant features from train set 
relevant_Features=['Safety_Score','Days_Since_Inspection','Accident_Type_Code','Violations']
for val in relevant_Features:
    sns.countplot(train[val])
    plt.show()
```

```python
train_drop = train.drop('Severity',axis = 1)
# Temperature in kelvin
train['Cabin_Temperature_kelvin'] = (train['Cabin_Temperature'] - 32) * 5/9 + 273.15
# log transformation
train['Turbulence_In_gforces_log'] = np.log(train['Turbulence_In_gforces'])
train['Adverse_Weather_Metric_log'] = np.log(train['Adverse_Weather_Metric'])
# Converting to Binary
train['Safety_Score_low'] =np.where(train['Safety_Score'] < 10, 1 , 0)
train['Days_Since_Inspection_high'] = np.where(train['Days_Since_Inspection'] > 20, 1 , 0)
train['Total_Safety_Complaints_high'] = np.where(train['Total_Safety_Complaints'] > 40, 1 , 0)
train['Max_Elevation_high'] = np.where(train['Max_Elevation'] > 45000, 1 , 0)
train['Max_Elevation_low'] = np.where(train['Max_Elevation'] < 5000, 1 , 0)
train['Adverse_Weather_Metric_high'] = np.where(train['Adverse_Weather_Metric'] > 2, 1 , 0)
train['Violations_high'] = np.where(train['Violations'] > 3, 1 , 0)
```

```python
#Removing irrelevant features
X = train.drop(['Severity','Turbulence_In_gforces','Accident_ID','Adverse_Weather_Metric','Total_Safety_Complaints','Max_Elevation','Cabin_Temperature','Violations'], axis=1)
X.head()
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import f1_score
lr = LogisticRegressionCV()
lr = LogisticRegressionCV(cv=2,class_weight="balanced", max_iter=7000,
                            n_jobs=-1,verbose=1,random_state=5,
                                penalty="l2",scoring="f1_weighted",solver="saga")

lr.fit(X,Y)
```

```python
lr.score(X,Y)
```

```python
#using gradient boosting classifier
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y = train.iloc[:, 0:1]
Y = encoder.fit_transform(Y['Severity'])
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingClassifier
from sklearn import metrics
classifier = GradientBoostingClassifier(n_estimators=1250,random_state=0,max_depth=6,min_samples_split=5,learning_rate=0.2)
classifier.fit(X,Y)
```

```python
GB_accuracies = cross_val_score(estimator=classifier,X=X,y=Y,cv=5,scoring='f1_weighted')
print("Mean GB Accuracies:",GB_accuracies.mean())
```
